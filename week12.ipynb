```python
import csv
f=open('seoul.csv')
data =csv.reader(f)
header = next(data)
max_diff=0
max_diff_date=''
max_temp=-1000
max_temp_date=''
min_temp=1000
min_temp_date=''
for row in data:
    if max_diff < abs(float(row[3]) - float(row[4])):
        max_diff=abs(float(row[3]) - float(row[4]))
        max_diff_date = row[0]
    if min_temp>float(row[3]):
        min_temp = float(row[3])
        min_temp_date = row[0]
    if max_temp<float(row[4]):
        max_temp = float(row[4])
        max_temp_date = row[0]
        
print("일교차가 가장큰날")
print(max_diff_date)
print("최저기온이 가장 낮은날")
print(min_temp_date)
print("최고기온이 가장높은날")
print(max_temp_date)
f.close()

f=open('subway.csv',encoding = 'utf-8')
data =csv.reader(f)
header = next(data)

line=[]
station_max=[]
station_min=[]

for row in data:
    if row[1] not in line:
        line.append(row[1])
        station_max.append([[0,''],[0,''],[0,'']])
        station_min.append([[1000000000000,''],[1000000000000,''],[1000000000000,'']])
    index = line.index(row[1])
    passa=float(row[4].replace(',',''))+float(row[5].replace(',',''))
    for i in range(3):
        if station_max[index][i][0]<passa:
            for j in range(2,i,-1):
                station_max[index][j], station_max[index][j-1] = station_max[index][j-1],station_max[index][j]
            station_max[index][i][0] = passa
            station_max[index][i][1]=row[3]
            break
    for i in range(3):
        if station_min[index][i][0]>passa:
            for j in range(2,i,-1):
                station_min[index][j], station_min[index][j-1] = station_min[index][j-1],station_min[index][j]
            station_min[index][i][0] = passa
            station_min[index][i][1]=row[3]
            break

for i in range(len(line)):
    print('\n======'+line[i] + '======')
    print('\n이용객이 가장 많은 역 순위\n')
    for j in range(3):
        print(str(j+1)+'.'+station_max[i][j][1]+'('+str(station_max[i][j][0])+')')
    print('\n이용객이 가장 적은 역 순위\n')
    for j in range(3):
        print(str(j+1)+'.'+station_min[i][j][1]+'('+str(station_min[i][j][0])+')')

```
