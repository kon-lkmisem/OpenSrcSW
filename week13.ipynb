```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

#실습1
f=open('seoul_2011-2020.csv')
data =csv.reader(f)
header = next(data)

date=[]
max_tem=[]
min_tem=[]

for row in data:
    if row[0].split("-")[1]=="08" and row[0].split("-")[2]=="11" and row[-1] !='' and row[-2] !='':
        date.append(int(row[0].split("-")[0]))
        max_tem.append(float(row[4]))
        min_tem.append(float(row[3]))

plt.title('8월11일의 연도별 최고기온 최저기온')
plt.plot(date,max_tem)
plt.plot(date,min_tem)
plt.show()

f.close()

#실습2

f=open('seoul_2011-2020.csv')
data =csv.reader(f)
header = next(data)
jan=[]
aug=[]
for row in data:
    if row[-1] !='':
        if row[0].split("-")[1]=='08':
            aug.append(float(row[-1]))
        elif row[0].split("-")[1]=='01':
            jan.append(float(row[-1]))

plt.title('서울 지역의 1월과 8월 최고기온')
plt.hist(aug,bins=100,color='r',label='Aug')  
plt.hist(jan,bins=100,color='b',label='Jan')
plt.legend()
plt.show()


```
