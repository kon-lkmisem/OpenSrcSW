```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

#실습1
f=open('seoul_all.csv')
data =csv.reader(f)
header = next(data)

data_all=[]
data_name=[]
data_g=[]
rank=[["",100000000000000000,0],["",0,0]]

for row in data:
    name=row[0].split(" ")[1]
    if name=="광진구":
        for i in range(3,104):
            data_g.append(int(row[i].replace(",","")))
    elif name!="":
        data_name.append(name)
        num=[]
        for i in range(3,104):
            num.append(int(row[i].replace(",","")))
        data_all.append(num)
        
for i in range(len(data_all)):
    x=0
    a=np.array(data_g)
    b=np.array(data_all[i])
    c=a-b
    x=np.std(c)
    if x<rank[0][1]:
        rank[0][0]=data_name[i]
        rank[0][1]=x
        rank[0][2]=i
    elif x>rank[1][1]:
        rank[1][0]=data_name[i]
        rank[1][1]=x
        rank[1][2]=i
print(rank)
plt.style.use('ggplot')
plt.plot(data_g,label="광진구")
plt.plot(data_all[rank[0][2]],label=rank[0][0]+"("+str(round(rank[0][1],2))+")")
plt.plot(data_all[rank[1][2]],label=rank[1][0]+"("+str(round(rank[1][1],2))+")")
plt.legend()
plt.show()
        
        
```
